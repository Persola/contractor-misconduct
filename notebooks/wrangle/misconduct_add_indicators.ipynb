```python
import re
import os

import pandas as pd
import pickle
```

```python
CLEAN_DATA_DIR = '../../data/clean/'

TRAIN_INPUT_FILENAME = f'{CLEAN_DATA_DIR}misconduct_train.pkl'
TEST_INPUT_FILENAME = f'{CLEAN_DATA_DIR}misconduct_test.pkl'
TRAIN_OUTPUT_FILENAME = f'{CLEAN_DATA_DIR}misconduct_indicators_train.pkl'
TEST_OUTPUT_FILENAME = f'{CLEAN_DATA_DIR}misconduct_indicators_test.pkl'
```

```python
def normalize_column_name(string):
    return re.sub('\s', '_', string).lower()
```

```python
with open(TRAIN_INPUT_FILENAME, 'rb') as input_file:
    misconduct = pickle.load(input_file)
```

```python
def add_indicators(input_filename, output_filename):
    with open(input_filename, 'rb') as input_file:
        misconduct = pickle.load(input_file)

    cols_to_indicate = [
        'misconduct_type',
        'court_type',
        'enforcement_agency',
        'contracting_party'
    ]

    for col_name in cols_to_indicate:
        misconduct = pd.concat(
            [
                misconduct,
                pd.get_dummies(
                    misconduct[col_name],
                    prefix=col_name
                )
            ],
            axis=1
        )

    misconduct = misconduct.drop(cols_to_indicate, axis=1)
    misconduct.columns = [
        normalize_column_name(col)
        for col
        in misconduct.columns
    ]
    
    os.system(f'touch {output_filename}')
    with open(output_filename, 'wb') as output_file:
        pickle.dump(misconduct, output_file)
```

```python
add_indicators(TRAIN_INPUT_FILENAME, TRAIN_OUTPUT_FILENAME)
add_indicators(TEST_INPUT_FILENAME, TEST_OUTPUT_FILENAME)
```
